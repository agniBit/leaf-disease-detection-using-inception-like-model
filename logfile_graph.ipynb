```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
if not os.path.exists('training_graphs'):
    os.makedirs('training_graphs')
with open("training_graphs/model_summry57.txt", "r") as ins:
    arr = []
    for line in ins:
        arr.append(line)
```

```python
dt = np.array(arr)
k = dt
```

```python
for i in range(0,len(dt) -1 ):
    print("{} {}".format(i,dt[i]))
```

```python
temp3 = dt[i].split(' ')
for i in range(0,len(dt) - 1):
    temp3 = dt[i].split(' ')
    print("{}  {}  {}  {}".format(temp3[5],temp3[9],temp3[15],temp3[-3]))
```

```python
E_P = []
V_A = []
T_A = []
V_L = []
T_L = []
for i in range(0,len(dt) - 1):
    temp3 = dt[i].split(' ')
    E_P.append( int(i) )
    T_L.append( float(temp3[5]) )
    V_L.append( float(temp3[9]))
    T_A.append( float(temp3[15]))
    V_A.append( float(temp3[-3]))
    print("{}    {}    {}    {}    {}".format( E_P[i], V_A[i], T_A[i], V_L[i], T_L[i]))
np.array( E_P )
np.array( T_L )
np.array( V_L )
np.array( T_A )
np.array( V_A )
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,T_A,lw = 1)
plt.xlabel('epoch')
plt.ylabel('training accuracy')
plt.title('training accuracy vs epochs')
plt.savefig('training_graphs/T_A_vs_epochs.png')
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,V_A,lw= 1)
plt.xlabel('epoch')
plt.ylabel('validation accuracy')
plt.title('validation accuracy vs epochs')
plt.savefig('training_graphs/V_A_vs_epochs.png')
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,V_L,lw= 1)
plt.xlabel('epoch')
plt.ylabel('validation loss')
plt.title('validation loss vs epochs')
plt.savefig('training_graphs/V_L_vs_epochs.png')
```

```python
plt.figure(figsize=(30,10))
plt.plot(E_P,T_L,lw= 1)
plt.xlabel('epoch')
plt.ylabel('training loss')
plt.title('training loss vs epochs')
plt.savefig('training_graphs/T_L_vs_epochs.png')
```

```python
z = np.polyfit(E_P, V_A, 3)


f = np.poly1d( z )
print( f )
lr  = np.zeros( 230 )
lr1 = np.zeros( 230 )
for i in range(0, 230):
    lr[i] = f(i)
    lr1[i] = i
```

```python
plt.figure(figsize=(30,10))

plt.plot(lr1, lr, lw = 3)
plt.xlabel('epoch')
plt.ylabel('validation accuracy')
plt.title('validation accuracy vs epochs best fir curve')
plt.savefig('training_graphs/V_A_vs_epochs_best_fir.png')
```
